# head() And tail() function

```python
dframe= read_csv("Students_data.csv")

new_df = dframe.head()   #Read only first five rows
new_df
```

```python
#If we paas the argument in the head() function then it will show the specified number of rows
new_df = dframe.head(2)
new_df
```

```python

new_df = dframe.tail()  # read only last 5 rows

new_df
```

```python
new_df = dframe.tail(3)  # if pass argument in tail function then it read specified rows from last
new_df
```
