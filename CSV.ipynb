# TO store csv file into DataFrame

```python
from pandas import *

dframe= read_csv(r"Students_data.csv")   #To read a csv file

dframe  # When you print a datframe, by default it will print only first five and last five rows 
```

```python
print(dframe.to_string())    # To print the entire dataframe
```

# Create New CSV file from existing file

```python
#Program to write Existing csv file into another 

dframe.to_csv("Students_new_data.csv")     #To create a new csv file from existing one

"""To check the newly created file from existing CSV file open the new(Students_new_data) file in excel or google sheets. """
```

```python
dframe.to_csv("Students_new_data.csv", columns=["Name","Roll_No"]) # Students_new_data.csv file contain only specified columns(Name and Roll_No)

"""To check the changes made to new.csv file open the file in excel."""
```

```python
dframe.to_csv("Students_new_data.csv", header=False)    #To remove the header from the new.csv file
```
