# Construct Dataframe in different ways

```python
#To create a dataframe from Python Dictionary

import pandas

weather= {'Day': ["1/1/2017","1/4/2017","1/5/2017","1/6/2017","1/7/2017","1/8/2017","1/9/2017","1/10/2017","1/11/2017"],
      "Temperature": [32,28,32,34,40,23,34,35,36],
      "Windspeed" : [6,9,7,8,12,12,13,14,10],
      "Event" : ["Rain","Sunny","Snow","Cloudy","Cloudy","Rain","Sunny","Snow","Sunny"]}

df= pandas.DataFrame(weather)   #Load the data into a dataframe object

df
```

```python
#To create dataframe from a list of tuples

import pandas as pd  #Import pandas as convention

weather = [
    ("1/1/2017", 32, 6, "Rain"),
    ("1/4/2017", 28, 9, "Sunny"),
    ("1/5/2017", 32, 7, "Snow"),
    ("1/6/2017", 34, 8, "Cloudy"),
    ("1/7/2017", 40, 12, "Cloudy"),
    ("1/8/2017", 23, 12, "Rain"),
    ("1/9/2017", 34, 13, "Sunny"),
    ("1/10/2017", 35, 14, "Snow"),
    ("1/11/2017", 36, 10, "Sunny")
    
]

df= pd.DataFrame(weather, columns=["Day", "Temperature", "Windspeed", "Event"])

df
```

```python
#To create a dataframe from list of dictionaries

weather = [
    {"Day" : "1/1/2017", "Temperature" : 32, "Windspeed" : 6, "Event" : "Rain"},
    {"Day" : "1/4/2017", "Temperature" : 28, "Windspeed" : 9, "Event" : "Sunny"},
    {"Day" : "1/5/2017", "Temperature" : 32, "Windspeed" : 7, "Event" : "Snow"},
    {"Day" : "1/6/2017", "Temperature" : 34, "Windspeed" : 8, "Event" : "Cloudy"},
    {"Day" : "1/7/2017", "Temperature" : 40, "Windspeed" : 12, "Event" : "Cloudy"},
    {"Day" : "1/8/2017", "Temperature" : 23, "Windspeed" : 12, "Event" : "Rain"},
    {"Day" : "1/9/2017", "Temperature" : 34, "Windspeed" : 13, "Event" : "Sunny"},
    {"Day" : "1/10/2017", "Temperature" : 35, "Windspeed" : 14, "Event" : "Snow"},
    {"Day" : "1/11/2017", "Temperature" : 36, "Windspeed" : 10, "Event" : "Sunny"}
]

df= pd.DataFrame(weather)

df
```

# Concat Dataframes

```python
from pandas import *

India_weather= DataFrame(
    {
    "City" : ["Mumbai", "Delhi", "Goa"],
    "Temp" : [32,45,20],
    "Humidity" : [60,65,79]
    }
)

America_weather= DataFrame(
    {
    "City" : ["Chicago", "London", "NewYork"],
    "Temp" : [24,37,23],
    "Humidity" : [65,75,61]
    }
)

new_df= concat([India_weather, America_weather])   #To join to dataframes
new_df
```

```python
new_df= concat([India_weather, America_weather], ignore_index= True)   #T0 get the continuous index
new_df
```

```python
# Pass keys to concat() function

new_df= concat([India_weather, America_weather], keys=["India", "America"])  # create additional index for the subset of dataframe
new_df
```

```python
# Concat dataframe column wise

Temp_df= DataFrame(
    {
    "City" : ["Mumbai", "Delhi", "Goa"],
    "Temp" : [32,45,20],
    }
)

Windspeed_df= DataFrame(
    {
    "City" : ["Goa", "Delhi", "Mumbai"],
    "windspeed" : [12,5, 9],
    }
)


new_df= concat([Temp_df, Windspeed_df], axis=1)   #To join to dataframes
new_df


#In output "City" column print twice 
```

```python
# If the order of cities in both dataframes is different then use the index argument

Temp_df= DataFrame(
    {
    "City" : ["Mumbai", "Delhi", "Goa"],
    "Temp" : [32,45,20],
    }, index=[0,1,2]
)

Windspeed_df= DataFrame(
    {
    "City" : ["Goa", "Delhi", "Mumbai"],
    "windspeed" : [12,5, 9],
    }, index=[2,1,0]
)

new_df= concat([Temp_df, Windspeed_df], axis=1)   #To join to dataframes
new_df
```
