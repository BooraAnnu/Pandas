# GroupBy (Split Apply Combine)

```python
from pandas import *

df = read_csv(r"GroupBy_data.csv")
df
```

```python
gp= df.groupby("City")   # group the dataframe by city 
gp
```

```python
# To access the data splitted by using groupby method

for city, city_df in gp:
    print(city)
    print(city_df)
    
```

```python
gp.get_group("Paris")   # Access the splitted data of city Paris only
```

```python
gp.max()   # get the maximum data of each city 
```

```python
gp.describe()   # Get all the statistics result of each city
```
